```python
Y = df.Salary
X = sm.add_constant(df.YearsExperience)
```

**SOME IMPORTANT EXPLANATIONS:**

***F-statistic:*** It doesn't matter much in this project, as we are working only on a simple linear regression project, hence with only one independent variable. This statistic tests if at least one independent variable has a significant effect on the dependent variable. Having only one independent variable, its significance can be directly assessed by the associated coefficient (in this case, "X") and its value.

***

**R-squared (R):** It's a statistical measure representing the proportion of variability in the dependent variable explained by the model. In this case, 95.7% of the variability in "Salary" is explained by the variability in "YearsExperience".

***

**Const:** It's the coefficient associated with the constant (intercept) of the regression model. In this case, the value is 24850 (2.485e+04).

***

**YearsExperience:** It's the coefficient associated with the independent variable "YearsExperience". It indicates the estimated change in "Salary" for a one-unit increase in "YearsExperience". In this case, 9449.96.

***

**P>|t| for Const and YearsExperience:** It represents the p-value associated with the t-test to test the null hypothesis that the coefficient is zero. If the p-value is less than significance level (usually 0.05), we can reject the null hypothesis and consider the coefficient statistically significant.

***

**Confidence Interval:** [0.025, 0.975] These are the lower and upper bounds of the confidence interval for the coefficients. In other words, with 95% confidence, we expect the true value of the coefficient to lie within this interval.

***

**Omnibus:** Test that assesses the normality of the residuals. A small value (and a high p-value) suggests that the residuals are approximately normal.

***

**Durbin-Watson:** It tests the autocorrelation of the residuals. The desired value is close to 2. Lower values suggest positive autocorrelation, and higher values suggest negative autocorrelation.

***

**Prob(Omnibus), Jarque-Bera (JB), Skew, Kurtosis:** Additional statistics that help diagnose the quality of the model fit.

***

**Cond.No:** Condition number, which measures the sensitivity of the model to small changes in the data. High values may indicate multicollinearity.

*Obtaining in-sample predictions*

```python
df['Salary_Predict'] = result_regression.predict()
```

```python
df.head()
```

Adding confidence interval

***
Usaremos 1.96 na conta do intervalo de confiana  95%. Pois para uma distribuio normal padro, aproximadamente 95% dos dados esto dentro de 1.96 desvios padro da mdia.
***
***
We will use 1.96 in the calculation of the 95% confidence interval. This is because for a standard normal distribution, approximately 95% of the data falls within 1.96 standard deviations from the mean.
***


```python
residual = df['Salary'] - df['Salary_Predict']
std_errors = residual.std()

df['Interval_Lower'] = df['Salary_Predict'] - 1.96 * std_errors
df['Interval_Upper'] = df['Salary_Predict'] + 1.96 * std_errors

df.head()
```

Scatter plot with model predictions

```python
YearsExperience = df['YearsExperience'].values
Salary = df['Salary'].values
Salary_Predict = df['Salary_Predict'].values
Interval_lower = df['Interval_Lower'].values
Interval_Upper = df['Interval_Upper'].values

plt.figure(figsize=(10,6))
plt.scatter(YearsExperience,Salary,label='Real',color='blue')
plt.scatter(YearsExperience,Salary_Predict, label='Foreseen',color='red')
plt.errorbar(YearsExperience,Salary_Predict,yerr=[Salary_Predict - Interval_lower, Interval_Upper - Salary_Predict],
            fmt='none',color='red',alpha=0.5,label = 'Cofidence Interval')

plt.title('Visual Comparison: Real vs. Foreseen',fontsize=20)
plt.xlabel('Years of Experience', fontsize=15)
plt.ylabel('Salary',fontsize=15)
plt.legend()

plt.show()
```

Observa-se com esse plot a diferena do Salrio real e do Salrio predito, percebemos que no h pomtos fora do comum. Conhece-se que as previses esto mais proximas da realidade para determinados intervalos de experincia.
***
***
With this plot, we observe the difference between the actual salary and the predicted salary. We can see that there are no outliers. It is noted that the predictions are closer to reality for certain experience intervals.

Residuals

```python
df['residuals'] = result_regression.resid
df.head()
```

Mean of the residuals

```python
df.residuals.mean()
```

```python
ax = sns.scatterplot(x=df.Salary_Predict,
                     y=df.residuals**2)
ax.figure.set_size_inches(12,6)
ax.set_title('Residual vs Predict Salary',fontsize=18)
ax.set_xlabel('Predict Salary', fontsize=14)
ax.set_ylabel('Residual',fontsize=14)
ax;
```

Atravs deste plot notamos que a varincia dos erros  constante.Significa que a disperso dos erros  uniforme ao longo de toda a faixa de valores da varivel explicativa.

Quando a constncia da varincia no  atendida, ou seja, h heterocedasticidade, os resduos podem mostrar padres de disperso que mudam com os valores da varivel independente. Isso pode resultar em problemas na estimativa dos parmetros do modelo e pode afetar a validade das inferncias estatsticas.
***
***
Through this plot, we notice that the variance of the errors is constant. This means that the dispersion of the errors is uniform across the entire range of values of the explanatory variable.

When the constancy of variance is not met, i.e., when there is heteroscedasticity, the residuals may exhibit patterns of dispersion that change with the values of the independent variable. This can result in problems in estimating the parameters of the model and may affect the validity of statistical inferences.
